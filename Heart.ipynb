```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\jason\Documents\Jerry\Studies\Jerry DSc\archive (2)\heart.csv")
```

```python
df.head()
```

```python
df['ChestPainType'].unique()
```

```python
df['ChestPainType'] = (
    df['ChestPainType'].replace(['ATA'],'0')
    .replace(['NAP'],'1')
    .replace(['ASY'],'2')
    .replace(['TA'],'3')
)
df.head()
```

```python
df['Sex'] = (
    df['Sex'].replace(['M'],'0')
    .replace(['F'],'1')
)
df.head()
```

```python
df['RestingECG'].unique()
```

```python
df['RestingECG'] = (
    df['RestingECG'].replace(['Normal'],'0')
    .replace(['ST'],'1')
    .replace(['LVH'],'2')
)
df.head()
```

```python
df['ExerciseAngina'] = (
    df['ExerciseAngina'].replace(['N'],'0')
    .replace(['Y'],'1')
)
df.head()
```

```python
df['ST_Slope'].unique()
```

```python
df['ST_Slope'] = (
    df['ST_Slope'].replace(['Up'],'0')
    .replace(['Flat'],'1')
    .replace(['Down'],'2')
)
df.head()
```

```python
df.describe()
```

```python
mean_value=df['Cholesterol'].mean()
mean_value
```

```python
mean_value=df['RestingBP'].mean()
mean_value
```

```python
print(df['RestingBP'].where(df['RestingBP'] == 0).count())
```

```python
print(df['Cholesterol'].where(df['Cholesterol'] == 0).count())
```

```python
df["Cholesterol"].replace(to_replace = 0, value = 198.80, inplace=True)
df["RestingBP"].replace(to_replace = 0, value = 132.40, inplace=True)
```

```python
df.describe()
```

```python
df.corr()
```

```python
y=df['HeartDisease']
y
```

```python
x=df.drop('HeartDisease', axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y)
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
from sklearn.linear_model import LogisticRegression
clr = LogisticRegression()
clr.fit(x_train, y_train)
y_pred = clr.predict(x_test)
y_pred
```

```python
print(confusion_matrix(y_test, y_pred))
accuracy = metrics.accuracy_score(y_test, y_pred)
print(accuracy)
print(metrics.classification_report(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(x_train, y_train)
y_pred2 = classifier.predict(x_test)
```

```python
print(confusion_matrix(y_test, y_pred2))
accuracy = metrics.accuracy_score(y_test, y_pred2)
print(accuracy)
print(metrics.classification_report(y_test, y_pred2))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
y_pred3 = dtc.predict(x_test)
y_pred3
```

```python
print(confusion_matrix(y_test, y_pred3))
accuracy = metrics.accuracy_score(y_test, y_pred3)
print(accuracy)
print(metrics.classification_report(y_test, y_pred3))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier_rf = RandomForestClassifier(random_state=42, n_jobs=-1, n_estimators=100)
classifier_rf.fit(x_train, y_train)
y_pred4 = classifier_rf.predict(x_test)
y_pred4
```

```python
print(confusion_matrix(y_test, y_pred4))
accuracy = metrics.accuracy_score(y_test, y_pred4)
print(accuracy)
print(metrics.classification_report(y_test, y_pred4))
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred5 = gnb.fit(x_train, y_train).predict(x_test)
y_pred5
```

```python
print(confusion_matrix(y_test, y_pred5))
accuracy = metrics.accuracy_score(y_test, y_pred5)
print(accuracy)
print(metrics.classification_report(y_test, y_pred5))
```
